```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
import matplotlib.pyplot as plt
import numpy as np

# 1. Load data
(train_img, train_labels), (test_img, test_labels) = tf.keras.datasets.fashion_mnist.load_data()

# 2. Pre-processing
# we need to reshape it for CNN because it needs 3 dimensions (28, 28, 1)
train_img = train_img.reshape(60000, 28, 28, 1)
train_img = train_img / 255.0

test_img = test_img.reshape(10000, 28, 28, 1)
test_img = test_img / 255.0

print("Data reshaped successfully.")

# 3. CNN Model
cnn = Sequential()

# convolution layer 1
cnn.add(Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)))
cnn.add(MaxPooling2D((2,2)))

# convolution layer 2
cnn.add(Conv2D(64, (3,3), activation='relu'))
cnn.add(MaxPooling2D((2,2)))

# fully connected layers
cnn.add(Flatten())
cnn.add(Dense(64, activation='relu'))
cnn.add(Dense(10, activation='softmax'))

cnn.summary()

cnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 4. Training
# using 8 epochs should be enough
h = cnn.fit(train_img, train_labels, epochs=8, validation_data=(test_img, test_labels))

# 5. Visualizing results
# accuracy
plt.plot(h.history['accuracy'])
plt.plot(h.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# loss
plt.plot(h.history['loss'])
plt.plot(h.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```
