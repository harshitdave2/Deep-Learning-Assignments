```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

# loading the mnist dataset
(xtrain, ytrain), (xtest, ytest) = tf.keras.datasets.mnist.load_data()

# checking shape just to be sure
print("Training data shape:", xtrain.shape)
print("Test data shape:", xtest.shape)

# normalizing the images to 0-1 range
xtrain = xtrain / 255.0
xtest = xtest / 255.0

# defining the model architecture
model = Sequential()
model.add(Flatten(input_shape=(28, 28)))  # input layer
model.add(Dense(128, activation='relu'))  # hidden layer
model.add(Dropout(0.2))
model.add(Dense(10, activation='softmax')) # output layer (0-9 digits)

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# training
print("\nStarting training...")
history = model.fit(xtrain, ytrain, epochs=10, validation_data=(xtest, ytest))

# plotting the accuracy graph
plt.plot(history.history['accuracy'], label='train acc')
plt.plot(history.history['val_accuracy'], label='val acc')
plt.title('Accuracy Plot')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# plotting loss graph
plt.figure() # new figure for loss
plt.plot(history.history['loss'], label='train loss')
plt.plot(history.history['val_loss'], label='val loss')
plt.title('Loss Plot')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# evaluation and confusion matrix
loss, acc = model.evaluate(xtest, ytest, verbose=0)
print(f"Final Test Accuracy: {acc*100:.2f}%")

# generating confusion matrix
ypred = model.predict(xtest)
ypred_classes = [np.argmax(element) for element in ypred]

cm = confusion_matrix(ytest, ypred_classes)
from sklearn.metrics import classification_report
# For Project 1
print(classification_report(ytest, ypred_classes))
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```
